```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "Salary": [4000, 4200, 4100, 80000, 4300, 4150]})
#IQR method of finding outliers
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
Lower=Q1-(1.5*IQR)
upper=Q3+(1.5*IQR)
outliers=df[(df<Lower)|(df>upper)]
print(outliers['Salary'].tolist())
```

```python
import pandas as pd
df = pd.DataFrame({
    "Age": [25, 30, -5, 1000, 40]})
valid=df[(df['Age']>=0) & (df['Age']<=120)]['Age']
middle=valid.median()
#print(middle)
#print(missing)
df['Age']=df['Age'].apply(lambda x: middle if x<0 or x>120 else x)
print(df)

```

```python
import pandas as pd
df = pd.DataFrame({
    "Income": [20000, 30000, 25000, 40000]})
min=df['Income'].min()
max=df['Income'].max()
normalised=(df['Income']-min)/(max-min)
print(normalised)
```

```python
import pandas as pd
df = pd.DataFrame({
    "Age": [5, 20, 67, 15, 40, 80]})
for Age in df['Age']:
    if Age<12:
        print('Child')
    elif 12 <= Age <=59:
        print('Adult')
    else:
        print('Senior')

```

```python
import pandas as pd
df = pd.DataFrame({
    "Voted": ["Yes", "No", "Yes", "No"]
})
df["Voted_numeric"] = df["Voted"].map({"Yes": 1, "No": 0})
print(df["Voted_numeric"])
```

```python
df = pd.DataFrame({
    "Date": ["2024-01-05", "2024-06-30"]
})
df["Year"] = pd.to_datetime(df["Date"]).dt.year
df['month']=pd.to_datetime(df['Date']).dt.month
df['day']=pd.to_datetime(df['Date']).dt.day
print(df["Year"])
print(df['month'])
print(df["day"])
```

```python
df = pd.DataFrame({
    "Country": ["Uganda", "Kenya"],
    "2020": [50, 60],
    "2021": [55, 65]
})
long_df = pd.melt(df, id_vars="Country", var_name="Year", value_name="Value")
print(long_df)
wide_df= long_df.pivot(index='Country', columns='Year', values='Value')
print(wide_df)
```

```python
df1 = pd.DataFrame({"Name": ["Alice", "Bob"]})
df2 = pd.DataFrame({"Name": ["Charles", "Diana"]})
df_combined = pd.concat([df1, df2])
print(df1)
print(df2)
print(df_combined)
```

```python
df1 = pd.DataFrame({"ID": [1, 2], "Name": ["Ali", "Becky"]})
df2 = pd.DataFrame({"ID": [1, 2], "Score": [90, 85]})
df_merged = pd.merge(df1, df2, on="ID")
print(df1)
print(df2)
print(df_merged)
```

```python
df1 = pd.DataFrame({"StudentID": [1, 2], "Name": ["Ali", "Becky"]})
df2 = pd.DataFrame({"ID": [1, 2], "Score": [90, 85]})
df_merged = pd.merge(df1, df2, left_on="StudentID", right_on="ID")
print(df1)
print(df2)
print(df_merged)
```
