```python
#implementing a stack
class Stack:
    def __init__(self):
        self.elements=[]

    def is_empty(self):
        if len(self.elements)==0:
            print('stack is empty')
        else:
            print(f'stack has {len(self.elements)} elements')
    
    def push(self, element):
        self.elements.append(element)
        print(f'{element} addded')

    def pop(self):
        if self.is_empty():
            print('stack is empty')
        else:
            print(f'{self.elements[-1]} removed')
            self.elements.pop()
            
    def peek(self):
        if self.is_empty():
            print('stack is empty')
        else:
            print(self.elements[-1])
    
    def display(self):
        print(self.elements[::-1])

    def search(self, key):
        key=int(input('enter element to search:'))
        if key in self.elements:
            #index_from_top=len(self.elements)-self.elements.index(key)
            id=reversed(self.elements.index(key))

            print(f'{key} found at index {id}')
        else:
            return 'element not found'
        
    def clear(self):
        self.elements=[]


stack=Stack()  
stack.push('x')
stack.push('y')
stack.push(6)
stack.push(7)
stack.pop()
stack.peek()
stack.display()
stack.is_empty()
stack.clear()




            
    
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class SinglyLinkedList:
    def __init__(self):
        self.head = None

    # i. Add node at end
    def add_end(self, data):
        new_node = Node(data)

        # If list is empty, the new node becomes the head
        if self.head is None:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # ii. Insert at a specific position (1-based index)
    def insert_at_position(self, data, position):
        new_node = Node(data)

        # Insert at the head
        if position == 1:
            new_node.next = self.head
            self.head = new_node
            return

        temp = self.head
        for _ in range(position - 2):
            if temp is None:
                print("Position out of range")
                return
            temp = temp.next

        new_node.next = temp.next
        temp.next = new_node

    # iii. Delete node by value
    def delete_by_value(self, value):
        if self.head is None:
            print("List is empty")
            return

        # If head is the node to delete
        if self.head.data == value:
            self.head = self.head.next
            return

        temp = self.head
        while temp.next and temp.next.data != value:
            temp = temp.next

        if temp.next is None:
            print("Value not found")
        else:
            temp.next = temp.next.next

    # iv. Display list
    def display(self):
        if self.head is None:
            print("List is empty")
            return

        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

    # v. Search by value and show position
    def search(self, value):
        temp = self.head
        position = 1

        while temp:
            if temp.data == value:
                print(f"Value {value} found at position {position}")
                return
            temp = temp.next
            position += 1

        print("Value not found")

    # vi. Get size of the list
    def size(self):
        temp = self.head
        count = 0
        while temp:
            count += 1
            temp = temp.next
        return count

    # vii. Delete first node
    def delete_first(self):
        if self.head:
            self.head = self.head.next

    # viii. Delete last node
    def delete_last(self):
        if self.head is None:
            return
        if self.head.next is None:
            self.head = None
            return

        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None

    # ix. Reverse the list
    def reverse(self):
        prev = None
        current = self.head

        while current:
            nxt = current.next
            current.next = prev
            prev = current
            current = nxt

        self.head = prev


# ------------------------------------
# Example usage
# ------------------------------------
ll = SinglyLinkedList()
ll.add_end(10)
ll.add_end(20)
ll.add_end(30)

print("Original List:")
ll.display()

ll.insert_at_position(15, 2)
print("After Inserting 15 at position 2:")
ll.display()

ll.delete_by_value(20)
print("After Deleting value 20:")
ll.display()

print("Searching for 30:")
ll.search(30)

print("Size of List:", ll.size())

ll.delete_first()
print("After Deleting First Node:")
ll.display()

ll.delete_last()
print("After Deleting Last Node:")
ll.display()

ll.reverse()
print("Reversed List:")
ll.display()
```

```python
from collections import deque

class queue:
    def __init__(self):
        self.items=deque()

    def is_empty(self):
        return len(self.items)
    
    def display(self):
        print(self.items)
        return self.items
    
    def enqueque(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.popleft()
    
    def peek(self):
        if not self.is_empty():
            print(self.items[0])

    def search(self, item):
        if not self.is_empty():
            id=self.items.index(item)
            if item in self.items:
                print(f'{item} found at index {id}')
            else:
                print('not found')

st=queue()
st.enqueque(5)
st.enqueque(6)
st.enqueque('w')
st.display()
st.dequeue()
```
